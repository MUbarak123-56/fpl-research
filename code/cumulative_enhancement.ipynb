```python
def reindex(year):
    file_new = file[file["season"] == year].reset_index(drop=True)
    file_new["date"] = pd.to_datetime(file_new["kickoff_time"]).dt.date
    file_new.index = file_new["date"]
    file_new = file_new.sort_index()
    date_range = pd.date_range(start=file_new.index.min(), end = file_new.index.max())
    file_new["name_team"] = file_new["name"] + "_" + file_new["team"]
    df = pd.DataFrame()
    for combo in list(file_new["name_team"].unique()):
        new_data = file_new[file_new["name_team"] == combo]
        post = new_data["position"].unique()[0]
        name, team = combo.split("_")[0], combo.split("_")[1]
        new_data_reindexed = new_data.reindex(date_range)
        new_data_reindexed["name"] = name
        new_data_reindexed["team"] = team
        new_data_reindexed["position"] = post
        new_data_reindexed["date"] = new_data_reindexed.index
        new_data_reindexed = new_data_reindexed.ffill(axis=0)
        new_data_reindexed = new_data_reindexed.dropna()
        new_data_reindexed = new_data_reindexed.reset_index(drop=True)
        df = pd.concat([df, new_data_reindexed], axis = 0).reset_index(drop=True)
    return df
```

```python
file2022_23 = reindex("2022/23")
```

```python
file2021_22 = reindex("2021/22")
```

```python
file2023_24 = reindex("2023/24")
```

```python
combo = pd.concat([file2021_22, file2022_23, file2023_24], axis = 0).reset_index(drop=True)
```
