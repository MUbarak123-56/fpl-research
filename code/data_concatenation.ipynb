```python
import pandas as pd
import janitor
import os
```

```python
os.listdir("../data/fpl_data")
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
for data in os.listdir("../data/fpl_data"):
    if ".csv" in data and "fpl" in data:
        new_df = pd.read_csv("../data/fpl_data/" + data, encoding="latin1")
        new_df["season"] = str(data).split(".")[0][-7:]
        df = pd.concat([df, new_df], axis = 0)
df = df.reset_index(drop=True)
```
